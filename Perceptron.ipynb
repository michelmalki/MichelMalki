```python
import pandas as pd
import numpy as np
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("forumlog-3-6_2024.csv")
df.head()
```

```python
df.describe()
```

```python
labeled_df = df[df["label"] != "spam"]
print(labeled_df)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import confusion_matrix, classification_report

data = labeled_df.drop(columns=["label", "author"])
target = labeled_df["label"]

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.8, random_state=42)

print(X_train)
print(y_train)
```

```python
print('shape of data', data.shape)
print('shape of target', target.shape)
```

```python
perceptron = Perceptron(max_iter=10, random_state=42)
perceptron.fit(X_train, y_train)
```

```python
predictor = perceptron.predict(X_test)
print(predictor)
```

```python
conf_matrix = confusion_matrix(y_test, predictor)
classification = classification_report(y_test, predictor)

# Print the confusion matrix and classification report
print("Confusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(classification)
```

```python
accuracy = (conf_matrix[0, 0] + conf_matrix[1, 1]) / conf_matrix.sum()
precision = conf_matrix[0, 0] / (conf_matrix[0, 0] + conf_matrix[0, 1])
recall = conf_matrix[0, 0] / (conf_matrix[0, 0] + conf_matrix[1, 0])
f1_score = 2 * (precision * recall) / (precision + recall)


print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall (Sensitivity): {recall:.4f}")
print(f"F1-Score: {f1_score:.4f}")
```
