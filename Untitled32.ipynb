```python
# Section 1: Imports and Data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Generate synthetic users
np.random.seed(42)
users = pd.DataFrame({
    'user_id': range(1, 101),
    'age': np.random.randint(22, 65, 100),
    'income': np.random.randint(30000, 150000, 100),
    'credit_score': np.random.randint(580, 850, 100),
    'debt': np.random.randint(0, 80000, 100),
    'risk_tolerance': np.random.randint(1, 6, 100)
})

# Section 2: KMeans Clustering for Behavior Types
features = users[['age', 'income', 'credit_score', 'debt']]
scaled = StandardScaler().fit_transform(features)
kmeans = KMeans(n_clusters=4, n_init=10, random_state=1)
users['cluster'] = kmeans.fit_predict(scaled)

# Visualize clusters
sns.pairplot(users, hue='cluster', vars=['age', 'income', 'credit_score'])
plt.show()

# Section 3: Content-Based Recommender
product_catalog = pd.DataFrame({
    'product_id': [1, 2, 3],
    'name': ['Secure Saver Plan', 'Growth ETF Bundle', 'Low-Interest Loan'],
    'risk_level': [1, 4, 2],
    'product_type': ['investment', 'investment', 'loan']
})

def recommend_product(user):
    profile_risk = user['risk_tolerance']
    diff = abs(product_catalog['risk_level'] - profile_risk)
    return product_catalog.loc[diff.idxmin()]

print("\n--- Sample Recommendations ---")
for _, user in users.head(5).iterrows():
    product = recommend_product(user)
    print(f"User {user['user_id']}: Recommend '{product['name']}'")

# Section 4: Portfolio Risk Balancing
def calculate_portfolio_risk(portfolio):
    weights = np.array(portfolio['allocation'])
    risk_scores = np.array(portfolio['risk_score'])
    return np.dot(weights, risk_scores)

sample_portfolio = pd.DataFrame({
    'investment': ['Bond A', 'ETF B', 'Crypto C'],
    'risk_score': [1, 3, 5],
    'allocation': [0.4, 0.4, 0.2]
})

print("\nPortfolio Risk Score:", calculate_portfolio_risk(sample_portfolio))

# Section 5: Loan Recommendation Scoring
# Binary target: likely to repay (1) or default (0) [synthetic]
users['will_repay'] = (users['credit_score'] > 650) & (users['debt'] < 40000)
users['will_repay'] = users['will_repay'].astype(int)

loan_model = LogisticRegression()
X = users[['age', 'income', 'credit_score', 'debt']]
y = users['will_repay']
loan_model.fit(X, y)

# Predict on new applicant
new_applicant = pd.DataFrame({
    'age': [32],
    'income': [60000],
    'credit_score': [700],
    'debt': [25000]
})
prob = loan_model.predict_proba(new_applicant)[0][1]
print("\nLoan Approval Probability:", round(prob * 100, 2), "%")
```

```python
# Step 1: Libraries & Synthetic User Data
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
users = pd.DataFrame({
    'user_id': range(1, 1001),
    'age': np.random.randint(20, 65, 1000),
    'income': np.random.normal(70000, 15000, 1000).astype(int),
    'risk_tolerance': np.random.randint(1, 6, 1000)
})

# Step 2: Randomly assign A/B groups
users['group'] = np.random.choice(['A', 'B'], size=len(users))

# Step 3: Define strategy rules (simplified)
def recommend_strategy(group):
    return 'Secure Saver Plan' if group == 'A' else 'Growth ETF Bundle'

users['product_recommended'] = users['group'].apply(recommend_strategy)

# Step 4: Simulate engagement (A lower CTR, B higher CTR)
users['engaged'] = users['group'].apply(
    lambda g: np.random.binomial(1, 0.30 if g == 'A' else 0.45)
)

# Step 5: Calculate group performance
summary = users.groupby('group')['engaged'].agg(['count', 'sum', 'mean'])
print("\nEngagement Summary:\n", summary)

# Step 6: Statistical Testing (Two-proportion z-test)
group_a = users[users['group'] == 'A']['engaged']
group_b = users[users['group'] == 'B']['engaged']

z_stat, p_value = stats.ttest_ind(group_a, group_b)
print(f"\nP-value from t-test: {p_value:.4f}")
if p_value < 0.05:
    print("🟢 Statistically significant difference between A and B")
else:
    print("🟡 No statistically significant difference")

# Step 7: Optional Visualization
sns.barplot(x='group', y='engaged', data=users)
plt.title("Engagement Rate by Strategy Group")
plt.ylabel("Engagement Rate")
plt.show()
```
