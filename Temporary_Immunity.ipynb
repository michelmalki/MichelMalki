```python
from random import random
from random import sample
import networkx as nx
import matplotlib.pyplot as plt
n = 20 # how many people (they will arrive one by one)
conn = 2 # how many connections each arriving person forms
G = nx.barabasi_albert_graph(n, conn)
L = nx.circular_layout(G)


perc = 0.07 # what percentage will be initially infected
sickPeople = round(perc * n)


infected = set(sample(list(G.nodes), sickPeople))
I = 'salmon'


contagion = 0.5 # probability of contagion

for test in range(5):
  if(random() < contagion):
   break
newlyInfected = set()
temporary_immunity = set()
spread = set()
newlyInfected = set()
for v in G.nodes:
  if v in infected:
    for u in G.neighbors(v): # set of nodes with which this one has a connection in the model
      if u not in infected and u not in temporary_immunity:
        if random() < contagion:
          spread.add((v, u))
          spread.add((u, v))
          newlyInfected.add(u)
len(spread)

newlyInfected = set()
temporary_immunity = set()
# cloring
Vs = [ I if v in infected else S for v in G.nodes ]
Es = [ I if (v, u) in spread else 'black' for v, u in G.edges ]

fig = plt.figure(1, figsize = (5, 5), dpi =  150)
nx.draw(G, L, node_color = Vs, edge_color = Es, node_size = 80)
l = nx.draw_networkx_labels(G, pos = L, font_size = 6)

temporary_immunity = 0.9
temporary_immunity = set()
for t in infected:
    if random() == temporary_immunity:
      temporary_immunity.add(t)
      temporary_immunity.remove(t)
    infected -= temporary_immunity
temporary_immunity

t = 'green'
Vs = [ I if v in infected else (t if v in temporary_immunity else 'green') for v in G.nodes ]
Es = [ I if (v, u) in spread else 'green' for v, u in G.edges ]
fig = plt.figure(1, figsize = (5, 5), dpi =  150)
nx.draw(G, L, node_color = Vs, node_size = 50)
l = nx.draw_networkx_labels(G, pos = L, font_size = 5)
recovery = 0.3 # probability of recovery
recovered = set()
for v in infected:
  if random() < recovery:
    recovered.add(v)
infected -= recovered
recovered

R = 'red'
Vs = [ I if v in infected else (R if v in recovered else S) for v in G.nodes ]
Es = [ I if (v, u) in spread else 'yellow' for v, u in G.edges ]
fig = plt.figure(1, figsize = (5, 5), dpi =  150)
nx.draw(G, L, node_color = Vs, node_size = 70)
l = nx.draw_networkx_labels(G, pos = L, font_size = 5)
```
