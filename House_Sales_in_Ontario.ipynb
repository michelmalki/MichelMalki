```python
def normalize(d):
  low = min(d)
  high = max(d)
  span = high - low
  return [ (value - low) / span for value in d]

n = 50

import pandas as pd
from random import randint


p = [ randint(450000, 5000000) for house in range(n) ]
price = pd.Series(p, name = 'Price')
nprice = pd.Series(normalize(p), name = 'NPrice')


lat = [ randint(41.0000000, 50.0000000)for house in range(n) ]
lat = pd.Series(lat, name = 'Latitude')

nlatitude = pd.Series(normalize(lat), name = 'NLatitude')

lg = [ randint(70.0000000, 90.0000000) for house in range(n) ]
longitude = pd.Series(lg, name = 'Longitude')
nlongitude = pd.Series(normalize(lg), name = 'NLongitude')


attribute = [price, nprice, lat, nlatitude, longitude, nlongitude]
house= pd.concat(attribute, axis= 1)
house.head(10)
```
