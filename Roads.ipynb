```python
import matplotlib.pyplot as plt
from random import random
import networkx as nx
from random import randint
import numpy as np


t = 20 # how many towns
p = nx.Graph() # province
for i in range(t):
  x = random()
  y = random()
  z = np.random.geometric(p=0.95, size=1000000)
  (z == 1).sum() / 1000000
  p.add_node(i, pos = (x, y), size = i)
coords = nx.get_node_attributes(p, 'pos')
pobs = nx.get_node_attributes(p, 'size')
pl = [pobs[n] for n in p.nodes()]
fig = plt.figure(1, figsize = (5, 5), dpi =  150)
nx.draw(p, coords, node_color = 'red', node_size = pl)
```

```python
from math import sqrt

def distance(p1, p2):
  (x1, y1) = p1
  (x2, y2) = p2
  return sqrt((x1 - x2)**2 + (y1 - y2)**2)

for t1 in p.nodes:
  l1 = p.nodes[t1]['pos']
  for t2 in p.nodes:
    if t1 != t2:
      l2 = p.nodes[t2]['pos']
      cost = distance(l1, l2) / (p.nodes[t1]['size'] + p.nodes[t2]['size'])
      p.add_edge(t1, t2, weight = cost)

nx.draw(p, coords, node_color = 'red', node_size = pl)
```

```python
r = nx.minimum_spanning_tree(p)
nx.draw(r, coords, node_color = 'red', node_size = pl)
```
