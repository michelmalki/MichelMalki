```python
import matplotlib.pyplot as plt
from random import random
import networkx as nx
from random import randint
import numpy as np


t = 35
G = nx.Graph()
for i in range(t):
  x = random()
  y = random()
  z = np.random.geometric(p=0.95, size=1000000)
  G.add_node(i, pos = (x, y), size = i)
coords = nx.get_node_attributes(G, 'pos')
m = nx.get_node_attributes(G, 'size')
pl = [m[n] for n in G.nodes()]
fig = plt.figure(1, figsize = (5, 5), dpi =  150)
nx.draw(G, coords, node_color = 'red', node_size = pl)
```

```python
from math import sqrt

def distance(p1, p2):
  (x1, y1) = p1
  (x2, y2) = p2
  return sqrt((x1 - x2)**2 + (y1 - y2)**2)

for t1 in G.nodes:
  l1 = G.nodes[t1]['pos']
  for t2 in G.nodes:
    if t1 != t2:
      l2 = G.nodes[t2]['pos']
      cost = distance(l1, l2) / (G.nodes[t1]['size'] + G.nodes[t2]['size'])
      G.add_edge(t1, t2, weight = cost)

nx.draw(G, coords, node_color = 'red', node_size = pl)
```

```python
r = nx.minimum_spanning_tree(G)
nx.draw(r, coords, node_color = 'red', node_size = pl)
```
