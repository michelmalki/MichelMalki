```python
import pandas as pd

df = pd.read_csv('house_sales.csv')
df['Country'] = 'Canada'
df.to_csv('updated_file.csv', index=False)
df.head()
```

```python
import matplotlib.pyplot as plt

df['price'].describe()
print(df['price'].describe())
plt.figure(figsize=(5, 3))
plt.hist(df['price'], bins = 25)
plt.title('Sale Price Distribution')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
print("Skewness after removing outliers:", df['price'].skew())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df['sqft_living'].describe()
print(df['sqft_living'].describe())
plt.figure(figsize=(5, 3))
plt.hist(df['sqft_living'], bins = 25)
plt.title('sqft_living Distribution')
plt.xlabel('sqft_living')
plt.ylabel('Density')
plt.show()
print("Skewness after removing outliers:", df['sqft_living'].skew())
```

```python
import matplotlib.pyplot as plt
df['bedrooms'].describe()
print(df['bedrooms'].describe())
plt.figure(figsize=(5, 3))
plt.hist(df['bedrooms'], bins= 25)
plt.title('bedrooms Distribution')
plt.xlabel('bedrooms')
plt.ylabel('Count')
plt.show()
print("Skewness after removing outliers:", df['sqft_living'].skew())
```

```python
import pandas as pd
x = df['price']
y = df['sqft_living']
plt.scatter(x,y, color = 'g')
plt.xlabel(' price')
plt.ylabel('sqft_living')
plt.show()
df = df[df['sqft_living']<7000]
```

```python
import pandas as pd
x = df['price']
y = df['zipcode']
plt.scatter(x, y, color = 'orange')
plt.xlabel(' price')
plt.ylabel('zipcode')
plt.title('Distribution of price and zipcode')
plt.show()
df = df[df['zipcode']<9801]
```

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.figure(figsize=(8, 4))
plt.scatter(df['yr_built'], df['price'], s=df['yr_renovated'] * 10, c=df['renovated'], cmap='coolwarm', alpha=0.7)
plt.xlabel('Property Age (Years Since Built)')
plt.ylabel('Sale Price')
plt.title('Impact of Year Built and Renovations on Sale Price')
plt.show()
```

```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(8, 4))
sns.heatmap(corrmat, vmax=.8, square=True);
```
