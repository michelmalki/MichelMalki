```python
# Analyze the distribution of key variables
#Identify outliers and understand their impact on the dataset.

import matplotlib.pyplot as plt
df['price'].describe()
print(df['price'].describe())
plt.figure(figsize=(5, 3))
plt.hist(df['price'], bins = 25)
plt.title('Sale Price Distribution')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
print("Skewness after removing outliers:", df['price'].skew())
```

```python
# Analyze the distribution of key variables
#Identify outliers and understand their impact on the dataset.

import matplotlib.pyplot as plt
import seaborn as sns
df['sqft_living'].describe()
print(df['sqft_living'].describe())
plt.figure(figsize=(5, 3))
plt.hist(df['sqft_living'], bins = 25)
plt.title('sqft_living Distribution')
plt.xlabel('sqft_living')
plt.ylabel('Density')
plt.show()
print("Skewness after removing outliers:", df['sqft_living'].skew())
```

```python
# Analyze the distribution of key variables
#Identify outliers and understand their impact on the dataset.

import matplotlib.pyplot as plt
df['bedrooms'].describe()
print(df['bedrooms'].describe())
plt.figure(figsize=(5, 3))
plt.hist(df['bedrooms'], bins= 25)
plt.title('bedrooms Distribution')
plt.xlabel('bedrooms')
plt.ylabel('Count')
plt.show()
print("Skewness after removing outliers:", df['sqft_living'].skew())
```

```python
#Explore the relationship between house prices and various features such as location (ZipCode), size (SqFtTotLiving), and property type (PropertyType)

import pandas as pd
x = df['price']
y = df['sqft_living']
plt.scatter(x,y, color = 'g')
plt.xlabel(' price')
plt.ylabel('sqft_living')
plt.show()
df = df[df['sqft_living']<7000]
```

```python
# Explore the relationship between house prices and various features such as location (ZipCode), size (SqFtTotLiving), and property type (PropertyType)

import pandas as pd
x = df['price']
y = df['zipcode']
plt.scatter(x, y, color = 'orange')
plt.xlabel(' price')
plt.ylabel('zipcode')
plt.title('Distribution of price and zipcode')
plt.show()
df = df[df['zipcode']<9801]
```

```python
# Investigate how features like the year built (YrBuilt) and renovations (YrRenovated) affect the sale price.

import matplotlib.pyplot as plt
import pandas as pd
plt.figure(figsize=(8, 4))
plt.scatter(df['yr_built'], df['price'], s=df['yr_renovated'] * 10, c=df['renovated'], cmap='coolwarm', alpha=0.7)
plt.xlabel('Property Age (Years Since Built)')
plt.ylabel('Sale Price')
plt.title('Impact of Year Built and Renovations on Sale Price')
plt.show()
```

```python
# Determine the variables that most strongly correlate with the SalePrice.

most_corr = df.corr()
most_corr = most_corr['price']
most_corr.sort_values(ascending=False)
```

```python
# Examine how house prices have changed over time.

plt.figure(figsize=(8, 4))
x=df['date']
y=df['price']
plt.scatter(x, y, color = 'r')
plt.xlabel('price')
plt.ylabel('year')
plt.title('Distribution of Prices Over Years')
plt.show()
```

```python
# Examine how house prices have changed over time.

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), cmap='BrBG', fmt='.2f', linewidths=2, annot=True)
plt.show()
```

```python
# Map the distribution of sale prices across different zip codes if possible.

import matplotlib.pyplot as plt
import pandas as pd
plt.figure(figsize=(10,6))
for zipcode in df['zipcode'].unique():
    # Filter data for each zip code
    data = df[df['zipcode'] == zipcode]
    plt.scatter([zipcode]*len(data), data['price'], alpha=0.5)

plt.title('Distribution of  Prices across Zip Codes')
plt.xlabel('Zip Code')
plt.ylabel(' Price')
plt.show()
```
