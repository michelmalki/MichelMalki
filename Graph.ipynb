```python
import networkx as nx
import matplotlib.pyplot as plt
i = 30
m = 1
F = nx.balanced_tree(i, m, create_using=None)
F1= F.number_of_nodes()
F2 = F.number_of_edges()
F1
F2
```

```python
D = nx.spectral_layout(F) # using NetworkX to visualize a graph (F)
fig = plt.figure(1, figsize = (5, 5), dpi = 120) # using spectral layout and drawing the graph with labels and red node color
nx.draw(F, D, with_labels = True,  node_color = 'red')
```

```python
from random import sample
# Assuming F is a NetworkX graph
d = sample(list(F.nodes), 2)
s1 = d[0]
t1 = d[1]
# Print nodes and edges
print("Nodes:", F.nodes())
print("Edges:", F.edges())
# Find the shortest path
path = nx.dijkstra_path(F, s1, t1)
path  #  the shortest path
```

```python
distance = nx.floyd_warshall(F)
distance[s1][t1]
```

```python
nx.average_shortest_path_length(F) # calculates the average shortest path length in the graph
```

```python
from random import choice
import networkx as nx

F5 = F.copy()  # Assuming F is a NetworkX graph
while True:
    edge_to_remove = choice(list(F.edges))
    F.remove_edge(*edge_to_remove)  # Unpack the edge tuple into two node arguments
    num11 = F.number_of_nodes()
    num22 = F.number_of_edges()
    print(f'{num11} vertices, down from {F1}') # vertices down from F5 which is number of nodes
    print(f'{num22} edges, down from {F2}') # edges down from F5 which is number of edges
    component = [e for e in nx.connected_components(F)]
    if len(component) > 1:
        break
    average = nx.average_shortest_path_length(F)
    print(f'Avg distance now at {average:.2f}')
# This snippet of code help the user to monitor the changes in the graph as edges are removed, and comparing them to the initial state of the graph (F5).
```
