```python
import networkx as nx
import matplotlib.pyplot as plt
i = 30
m = 1
F = nx.balanced_tree(i, m, create_using=None)
F1= F.number_of_nodes()
F2 = F.number_of_edges()
F1
F2
```

```python
D = nx.spectral_layout(F)
fig = plt.figure(1, figsize = (5, 5), dpi = 120)
nx.draw(F, D, with_labels = True,  node_color = 'red')
components = nx.connected_components(F)
```

```python
from random import sample
d = sample(list(F.nodes), 2)
s1 = d[0]
t1 = d[1]
print(F, 0 in F.nodes)
path = nx.dijkstra_path(F, s1, t1)
path
```

```python
distance = nx.floyd_warshall(F)
distance[s1][t1]
```

```python
nx.average_shortest_path_length(F)
```

```python
from random import choice
F4 = F.copy()
while True:
  broken = choice(list(F.nodes))
  F.remove_node(broken)
  num1 = F.number_of_nodes()
  num2 = F.number_of_edges()
  print(f'{num1} vertices, down from {F1}')
  print(f'{num2} edges, down from {F2}')
  component = [ e for e in nx.connected_components(F) ]
  if len(component) > 1:
    break
  average = nx.average_shortest_path_length(F)
  print(f'Avg distance now at {average:.2f}')
```

```python
percolated = r.copy()
import seaborn as sns

fade = set()
growth = [] # let's keep track of the avg distance while we hammer away

def start():
  nx.draw(r, P, node_size = 25)

def percolate(t):
  global percolated, fade
  comp = [ r for r in nx.connected_components(percolated) ]
  ccount = len(component)
  if ccount == 1: # still in one piece
    growth.append(nx.average_shortest_path_length(percolated))
  r5 = sns.color_palette('bright', ccount)

  Vc = [ 'lightgreen' for v in r.nodes ]
  i = 0
  for e in component:
    compColor = r5[i]
    for v in e:
      Vc[v] = r5[i]
    if i < len(r5) - 1:
      i += 1
  if percolated.number_of_nodes() > 1:
    v = choice(list(percolated.nodes))
    Vc[v] = 'lightgreen'
    for u in r.neighbors(v):
      fade.add((u, v))
      fade.add((v, u))
    percolated.remove_node(v)
  Ec = [ 'lightgreen' if e in fade else 'black' for e in r.edges() ]
  plt.clf()
  nx.draw(r, P, node_size = 25, node_color = Vc, edge_color = Ec)
  fig = plt.figure(figsize = (12, 5), dpi =  100)
```

```python
import matplotlib.pyplot as plt
len(growth)
if growth:
    low = min(growth)
    high = max(growth)
    margin = 0.05 * (high + low) / 2
    fig = plt.figure(figsize=(12, 5), dpi=100)
    l = len(growth)
    plt.xlim(-0.5, l - 0.5)
    plt.ylim(low - margin, high + margin)
    t = [i for i in range(l)]
    plt.plot(t, growth, c='grey', marker='o')
    plt.xlabel('Amount of vertices removed')
    plt.ylabel('Average distance')
    plt.show()
else:
    print("The 'growth' list is empty.")
```
